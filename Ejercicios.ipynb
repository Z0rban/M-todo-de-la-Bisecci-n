# TAREA METODO DE BISECCIÓN

## Algoritmo metodo de bisección

```python
from typing import Callable
def sign(x: float) -> int:
    if x > 0:
        return 1
    elif x < 0:
        return -1
    else:
        return 0
```

```python
def biseccion(
    a: float, b: float, *, ecuacion: Callable[[float], float], tol: float, N: int
) -> tuple[float, float, float, int] | None:
    i = 1

    assert a < b, "a no es menor que b, el intervalo no es válido."

    assert (
        ecuacion(a) * ecuacion(b) < 0
    ), "La función no cambia de signo en el intervalo."

    Fa = ecuacion(a)
    p = a 
    for i in range(N):
        p = a + (b - a) / 2
        FP = ecuacion(p)
        if FP == 0 or (b - a) / 2 < tol:
            return p, a, b, i

        if sign(Fa) * sign(FP) > 0:
            a = p
            Fa = FP
        else:
            b = p

    return p, a, b, i
```

## EJERCICIO 1
$$ x^3 − 7x^2 + 14x − 6  $$

```python
def ecuacion(x:float)->float:
    return (x**3)-(7*x**2)+14*x-6
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 300)

y = ecuacion(x)

plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de $x^3 + 7x^2 + 14x - 6$')
ax = plt.gca()
ax.set_ylim([-2, 4])
ax.set_xlim([0, 5])
plt.grid(True)
plt.legend(["$y(x)=x^3+7x^2+14x-6$"])
plt.show()
```

### Intervalo [0,1]

```python
a = 0
b = 1
tolerancia = 1e-3
max_iterations = 100

biseccion(a, b, ecuacion=ecuacion, tol=tolerancia, N=max_iterations)
```

### Intervalo [1,3.2]

```python
a = 1
b = 3.2
tolerancia = 1e-3
max_iterations = 100

biseccion(a, b, ecuacion=ecuacion, tol=tolerancia, N=max_iterations)
```

## EJERCICIO 4
$$ y = x^2 − 1 $$ 
$$ y = e^{1-x^2} $$
## Gráficas

```python
import math
e=math.e
def ecuacion1(x:float)->float:
    return (x**2)-1
def ecuacion2(x:float)->float:
    return e**(1-x**2)
```

```python
x = np.linspace(-10, 10, 300)

y1 = ecuacion1(x)
y2 = ecuacion2(x)
plt.plot(x, y1, label="$y(x)=x^2 -1$")
plt.plot(x, y2, label="$y(x)=e^{1-x^2}$")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de $ x^2 -1$  y  $ e^{1-x^2}$')
ax = plt.gca()
ax.set_ylim([-2, 5])
ax.set_xlim([-4, 4])
plt.legend()
plt.grid(True)
plt.show()
```

### Literal b

```python
def ecuacion_1_2(x:float)->float:
    return (x**2)-1 - e**(1-x**2)
```

#### Gráfica

```python
x = np.linspace(-10, 10, 300)

y = ecuacion_1_2(x)
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de $ x^2 -1 - e^{1-x^2}$')
ax = plt.gca()
ax.set_ylim([-2, 5])
ax.set_xlim([-4, 4])
plt.legend('$ x^2 -1 - e^{1-x^2}$')
plt.grid(True)
plt.show()
```

#### Metodo de Bisección

```python
a = -2
b = 0
tolerancia = 1e-3
max_iterations = 100

biseccion(a, b, ecuacion=ecuacion_1_2, tol=tolerancia, N=max_iterations)
```
